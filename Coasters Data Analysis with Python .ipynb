## Data Preparation

####  -Dropping insignificant rows and columns
####  -Identifying duplicats
####  -Renaming columns

```python
df=df[['coaster_name','Location','Status','Manufacturer','year_introduced','latitude','longitude', \
       'Type_Main','opening_date_clean','speed_mph','height_ft','Inversions_clean','Gforce_clean']].copy()
```

```python
df.columns
```

```python
df=df.rename(columns={'coaster_name':'Coaster_Name',
           'year_introduced':'Year_Introduced',
          'longitude':'Longitude',
          'opening_date_clean':'Opening_Date_Clean',
          'speed_mph':'Speed_mph',
          'height_ft':'Height_ft',
          'Inversions_clean':'Inversions',
          'Gforce_clean':'Gforce'})
```

## What are the locations with the avg fastest roller coasters (minimum of 10)?


```python
ax= df.query(" Location != 'Other' ").groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10') \
.sort_values('mean')['mean'].plot(kind='barh',title='Average Coast Speed by Location',figsize=(12,5))

ax.set_xlabel('Average Coaster Speed')
```
